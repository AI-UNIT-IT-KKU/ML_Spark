## **Training**: The ML algorithm processes this data and "trains" a model by adjusting its internal parameters to recognize the patterns that connect the inputs (Size of hous, No. of rooms) to the outputs (prices).

## **Prediction**: Once trained, the model can be given new, unseen data (a new house data) and make an accurate prediction (estimate its price).

## **Main Types of Machine Learning**

* Supervised Learning: The model learns from data that is labeled with the "right answers." Example: Predicting house prices using a dataset of houses that includes their final sale price.

* Unsupervised Learning: The model finds hidden structures and patterns in unlabeled data. Example: Grouping customers into different segments based on their purchasing habits, without any predefined labels.

# Common Symbols in Machine Learning

## Common Symbols in Machine Learning



| Symbol | Description & Usage |
| :--- | :--- |
| **Data Representation** | |
| $m$ | **Number of training examples** in your dataset. |
| $x$ | A single **input example**, typically a vector of all features. |
| $y$ | The **true label** or actual value for an input example. |
| $x^{(i)}$ | The features for the **i-th training example**. |
| $y^{(i)}$ | The true label for the **i-th training example**. |
| $x_{j}^{(i)}$ | The value of the **j-th feature** in the **i-th training example**. |
| **Model & Predictions** | |
| $w, b$ | The **parameters** of the model. **`w`** are the weights and **`b`** is the bias. These are the values the model learns. |
| $\hat{y}$ | The **prediction** or hypothesis made by the model. Pronounced "y-hat". |
| $f_{w,b}(x)$ | The **model's function** that maps an input `$x$` to a prediction `$\hat{y}$`. |
| $\sigma(z)$ | The **Sigmoid function**. Used in Logistic Regression to squash a value `$z$` into a probability between 0 and 1. |
| **Cost Function** | |
| $J(w,b)$ | The **Cost Function**. It measures the average error of the model's predictions. The goal of training is to minimize this value. |
| $\frac{\partial J}{\partial w}$ | The **gradient** of the cost function. It measures the slope of the cost function and tells Gradient Descent which direction to move in. |
| **Hyperparameters** | |
| $\alpha$ | The **Learning Rate**. This hyperparameter controls how large of a step Gradient Descent takes during each iteration. A small `$\alpha$` leads to slow but stable learning, while a large `$\alpha$` can cause the algorithm to "overshoot" the optimal solution. |
| Threshold | A value used in classification to convert a probability `$\hat{y}$` into a final class prediction. **Usage:** If `$\hat{y}$` >= Threshold (e.g., 0.5), predict class 1; otherwise, predict class 0. |

## 2.2 Categorical Data Encoding

When working with categorical data, we need to convert it into **numeric form** before using it in regression or machine learning models. Two common methods are **Label Encoding** and **One-Hot Encoding**.

---

** 1. Label Encoding **
- Converts each category into a **unique integer** in only ne column.
- Example: Target variable (Pass/Fail) → Pass = 1, Fail = 0.
- Good for the **target variable (y)** because regression/classification algorithms only need a numeric label for the output.
- Not always suitable for features, since it introduces a **false order** between categories.

---

** 2. One-Hot Encoding (Dummy Variables) **
- Creates a new binary column (0/1) for each category.
- Example: Region (North, South, East):

| Obs | Region |
|-----|--------|
| 1   | North  |
| 2   | South  |
| 3   | East   |


| Obs | North | South | East |
|-----|-------|-------|------|
| 1   | 1     | 0     | 0    |
| 2   | 0     | 1     | 0    |
| 3   | 0     | 0     | 1    |

- Best for **features (X)** because it removes the false ordering problem.
- Issue: can lead to the **Dummy Variable Trap**.

---

** 3. The Problem: Dummy Variable Trap **
- If we create a dummy for **every category**, the dummies become **perfectly collinear**.
- In the example above:
  
\[
North + South + East = 1
\]

- This causes **perfect multicollinearity**, which makes regression fail.

---

** 4. The Solution**
- Use only **k − 1 dummies** for \(k\) categories.
- Drop one category → it becomes the **baseline (reference group)**.

Example (drop East as baseline):

| Obs | North | South |
|-----|-------|-------|
| 1   | 1     | 0     |
| 2   | 0     | 1     |
| 3   | 0     | 0     | ← Baseline = East  

**Interpretation:**  
- Coefficients for North and South are measured **relative to East**.

## 2.3 Feature Selection

The process of choosing the most relevant predictors (features) for a machine learning model.  

### 1. Methods

- **All-in** → include all predictors (when theory justifies).  
- **Backward Elimination** → start with all predictors, remove the least significant (based on p-value).  
- **Forward Selection** → start empty, add features one by one if significant.  
- **Bidirectional (Stepwise)** → combine forward and backward steps.  
- **Score Comparison** → compare models with metrics like R², Adjusted R², AIC, BIC, CV accuracy.

---

## 2. Correlation Measures

Correlation quantifies the **strength and direction** of the relationship between variables.

### 2.1 Pearson Correlation
- Measures **linear** relationships.  
- Range: -1 to +1.  
- Limitation: fails with non-linear patterns (e.g., sine wave, U-shape).  

image.png


### 2.2 Spearman Correlation

- **Definition:**  
  Spearman correlation is **rank-based**.  
  It measures how well the relationship between two variables can be described by a **monotonic function**.  

- **Range:**  
  \(-1\) to \(+1\)  
  - \(+1\): perfect increasing monotonic relationship  
  - \(-1\): perfect decreasing monotonic relationship  
  - \(0\): no monotonic relationship  
  

### 2.3 Xi Correlation (ξ)
- Captures **any type of dependency** (linear, non-linear, monotonic, complex).  
- Range: 0 (independent) to 1 (perfect dependence).  
- More powerful for feature selection and data exploration.  

image.png

## 3. Statistical Significance (p-value)
Correlation (or regression coefficients) must be tested for **significance**.  

- **Null hypothesis (H₀):** no relationship (true correlation = 0).  
- **Alternative (H₁):** there is a relationship.  

### 🔹 p-value
- **p < 0.05** → strong evidence against H₀ → significant relationship.  
- **p > 0.05** → weak evidence → the relationship may not be real.  

- Correlation coefficient tells **how strong** the relationship is.  
- p-value tells **whether it is statistically reliable**.  



```
import pandas as pd
import statsmodels.api as sm

# Salary predicted by Age and Gender (0 = Female, 1 = Male)
data = {
    "Age": [22, 25, 30, 35, 40, 45, 50],
    "Gender": [0, 1, 0, 1, 0, 1, 0],
    "Salary": [2500, 2700, 3200, 3500, 4000, 4300, 4800]
}

df = pd.DataFrame(data)

# 🔹 Define predictors (X) and target (Y)
X = df[["Age", "Gender"]]
X = sm.add_constant(X)   # add intercept
y = df["Salary"]

# 🔹 Fit regression model
model = sm.OLS(y, X).fit()

# 🔹 Show regression summary
print(model.summary())

```

- If CI does not include 0 → the predictor is significant.

# 2.4 Feature Scaling


Goal: To bring all numerical features to a common scale. This is important because many algorithms are sensitive to the scale of the data, and features with larger ranges can unfairly dominate the training process.

⚠️ Critical Rule: You must fit the scaler only on the training data (X_train) and then use it to transform both the training and testing data (X_train and X_test). This prevents information from the test set from "leaking" into your training process.


Min-Max Scaling	$$X_{\text{norm}} = \frac{X - X_{\min}}{X_{\max} - X_{\min}}$$


Standardization	$$X_{\text{stand}} = \frac{X - \mu}{\sigma}$$

Method 1: Normalization (Min-Max Scaling)
This method rescales the data to a fixed range, usually 0 to 1.



```
# Normalization
from sklearn.preprocessing import MinMaxScaler

# 1. Initialize the scaler
scaler = MinMaxScaler()

# 2. Fit on the training data and transform it
X_train = scaler.fit_transform(X_train)

# 3. Transform the testing data using the same scaler
X_test = scaler.transform(X_test)
```



Method 2: Standardization (Z-score)
This method transforms the data to have a mean of 0 and a standard deviation of 1.

````python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers=[('scaler', StandardScaler(), [2, 3])],
    remainder='passthrough'
)

# Fit on training data and transform it
X_train_scaled = ct.fit_transform(X_train)

# Transform testing data with same scaler
X_test_scaled = ct.transform(X_test)
````

**We do not call fit on the test set because the test data must be transformed using the same scaling parameters (mean, std) that were learned from the training set.**

**If we fit on the test set, we would leak information about it into the model, which breaks the purpose of having unseen data for evaluation.**
